```python
### importing liberary
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import joblib
```

```python
## importing data
data =pd.read_csv('D:/Lecture file/25 th lecture 29 march/Decision Tree  - Assignment -4/iphone_purchase_records.csv')
data.head()
```

```python
data['Gender'].unique()
```

```python
data['Gender']=data['Gender'].replace({'Male': 0,'Female': 1})
data.head()
```

```python
data.info()
```

```python
print(f' Data null value total : \n {data.isnull().sum()}')
```

```python
X = data[['Gender','Age','Salary']]
y =data[['Purchase Iphone']]
```

```python
 ## importing   liberary fro descision tree 

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import  accuracy_score
```

```python
Dt = DecisionTreeClassifier(criterion='entropy')
```

```python
Dt.fit(X,y)
```

```python
fimp = pd.Series(Dt.feature_importances_,index=X.columns)
fimp.plot(kind ='barh')
plt.savefig('D:/series bar.png')
```

```python
pred_Dt =Dt.predict(X)
pred_Dt
```

```python
print(f'The Accuracy of the purchase prediction model on Decision tree is: {accuracy_score(y,pred_Dt)*100}%')
```
